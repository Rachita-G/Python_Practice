```python
import requests
```

Extracting all the urls from one webpage

```python
# link of the page you want to scrape
url="https://www.analyticsvidhya.com/blog/2017/07/data-visualisation-made-easy/"
```

```python
# get the webpage creating a response variable
response=requests.get(url)
```

```python
response 
# here 200 means your url is correct, internet connection works and you are allowed to access this page 
```

```python
# extract source code of the page into memory
data = response.text
```

```python
data
```

```python
soup= BeautifulSoup(data,'html.parser')
```

```python
soup
```

```python
# to extract specific html 'a' tags
tags=soup.find_all('a')
```

```python
for tag in tags:
    print(tag.get('href'))
```
